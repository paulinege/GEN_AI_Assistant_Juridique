## **1.Ajout d'une mémoire conversationnelle.**
L'objectif est de permettre à l'assistant de **conserver le contexte sur plusieurs tours**.

### **Idée générale**
Une conversation est identifiée par un `thread_id`.
Tant qu'on réutilise ce même `thread_id`, l'agent peut retrouver ce qui a été dit auparavant.

## **1.1 Préparation de l'environnement**

Comme le notebook est dans `notebooks/`, on ajoute la racine du projet au `sys.path`.

```python
import sys
import os
from pathlib import Path
from dotenv import load_dotenv

PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

ENV_PATH = PROJECT_ROOT / ".env"
print("PROJECT_ROOT =", PROJECT_ROOT)
print(".env existe ?", ENV_PATH.exists())

load_dotenv(ENV_PATH, override=True)

key = os.getenv("OPENAI_API_KEY")
print("OPENAI_API_KEY chargée ?", bool(key))
print("Préfixe :", key[:12] if key else None)
print("Suffixe :", key[-6:] if key else None)
```

## **1.3. Démonstration d'une conversation avec mémoire**

On crée un `thread_id` unique pour la session.
Ensuite, on envoie plusieurs messages dans **la même conversation**.

```python
thread_id = new_thread_id("juridique")
print("Thread ID :", thread_id)
```

```python
thread_id = new_thread_id("juridique")

print(ask_agent_with_memory("Bonjour, je m'appelle Sarah.", thread_id=thread_id))
print(ask_agent_with_memory("Je travaille sur un dossier de licenciement.", thread_id=thread_id))
print(ask_agent_with_memory("Quel est mon prénom ?", thread_id=thread_id))
print(ask_agent_with_memory("Sur quel type de dossier je travaille ?", thread_id=thread_id))
```
